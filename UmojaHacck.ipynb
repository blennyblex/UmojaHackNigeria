```python
features = list(X.columns)
categorical_indices = np.where(X.dtypes != np.float)[0]
def metric(y, pred):
    return f1_score(y, pred, labels=[0,1])
```

```python



n_skf = 12
score_list = []
score=0
pred_test = []

from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=n_skf)

skf.get_n_splits(X, y)
print(skf)

i = 0
for train_index, test_index in skf.split(X, y):
#     print("TRAIN:", train_index, "TEST:", test_index)
    X_train, X_test = X.loc[train_index], X.loc[test_index]
    y_train, y_test = y.loc[train_index], y.loc[test_index]
    
    
    model_cb = CatBoostClassifier(n_estimators = 4000,cat_features = categorical_indices, random_seed = 2020,
                                  learning_rate = 0.1, max_depth = 10,
                              eval_metric = "AUC", verbose = 100 , objective ="CrossEntropy",
                                  early_stopping_rounds = 1000, use_best_model = True)
                                
    
    model_cb.fit(X_train, y_train, eval_set = (X_test, y_test), use_best_model = True, verbose = 100)
    
    
    pred = model_cb.predict(X_test)
    
    sc = f1_score(y_test, pred)
    score_list.append(sc)
    score+= sc/n_skf
    
    i+= 1
    
    predicts = model_cb.predict(test[features])
    pred_test.append(predicts)
    print('Fold {} : {}'. format(i, sc))
    
print()
print()
print('Average log :', score)    
```
